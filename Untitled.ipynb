```python
# -*- coding: utf-8 -*-
"""
Created on Wed Dec  2 20:36:04 2020

@author: SONG XU
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
        'Google text to speech':['95','87','82','84', '88'],
        'pyttsx3':['62','75','64','68','76'],
        }
df = pd.DataFrame(data,index=['iPad','iPhone','MacBook', 'Dell XPS', 'Huawei'],
               columns=['Google text to speech', 'pyttsx3'])
```

```python
df
```

```python
plt.figure(figsize=(10,10))
ax = sns.barplot(x=df.index, y='Google text to speech', data=df);
ax.set_xticklabels(ax.get_xticklabels(),fontsize=20);
plt.title("Google text to speech performance", fontsize=25)
```

```python
plt.figure(figsize=(10,10))
ax = sns.barplot(x=df.index, y='pyttsx3', data=df);
ax.set_xticklabels(ax.get_xticklabels(),fontsize=20);
plt.title("pyttsx3 performance",fontsize=25)
```
